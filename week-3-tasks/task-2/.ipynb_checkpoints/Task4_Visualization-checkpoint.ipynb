## Import & Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

df = pd.read_csv("Student_performance_data.csv") 
```

#### GPA Distribution (Histogram)

```python
plt.figure(figsize=(8,5))
sns.histplot(df['GPA'], bins=20)
plt.title("GPA Distribution")
plt.savefig("1_gpa_distribution.png", dpi=300, bbox_inches='tight')
plt.show()
```

#### GPA by Gender (Boxplot)

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Gender', y='GPA', data=df)
plt.title("GPA by Gender")
plt.savefig("2_gpa_gender.png", dpi=300, bbox_inches='tight')
plt.show()
```

#### Study Time vs GPA

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='StudyTimeWeekly', y='GPA', data=df)
plt.title("Study Time vs GPA")
plt.savefig("3_study_vs_gpa.png", dpi=300, bbox_inches='tight')
plt.show()
```

#### Absences vs GPA

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Absences', y='GPA', data=df)
plt.title("Absences vs GPA")
plt.savefig("4_absences_vs_gpa.png", dpi=300, bbox_inches='tight')
plt.show()
```

#### Grade Class Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(x='GradeClass', data=df)
plt.title("Grade Class Distribution")
plt.savefig("5_gradeclass.png", dpi=300, bbox_inches='tight')
plt.show()
```

#### Correlation Heatmap

```python
plt.figure(figsize=(10,6))
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.savefig("6_heatmap.png", dpi=300, bbox_inches='tight')
plt.show()
```

#### Extracurricular vs GPA

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Extracurricular', y='GPA', data=df)
plt.title("Extracurricular Activities vs GPA")
plt.savefig("7_extra_vs_gpa.png", dpi=300, bbox_inches='tight')
plt.show()
```

### Insights

### GPA is normally distributed among students
### Study time shows a positive relationship with GPA
### Higher absences are associated with lower GPA
### Extracurricular activities show moderate impact on performance
### Gender-based GPA differences are minimal
### Correlation heatmap highlights relationships between numerical features
