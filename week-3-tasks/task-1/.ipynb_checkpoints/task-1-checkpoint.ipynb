## Importing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

### Loading dataset

```python
df = pd.read_csv("Student_performance_data.csv")
```

```python
df.sample(5)
```

### Shape + Columns

```python
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
```

```python
df.info()
```

### Missing Values

```python
plt.figure(figsize=(10,5))
df.isnull().sum().plot(kind='bar')
plt.title("Missing Values Overview")
plt.savefig("task1_missing.png", dpi=300, bbox_inches='tight')
plt.show()
```

### Data Types Distribution

```python
df.dtypes.value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Data Types Distribution")
plt.savefig("task1_dtypes.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
df.describe()
```

```python
df.columns
```

### Target Variable Understanding

```python
plt.figure(figsize=(8,5))
sns.histplot(df['GPA'], bins=20)
plt.title("GPA Distribution")
plt.savefig("task1_gpa_distribution.png", dpi=300, bbox_inches='tight')
plt.show()
```

### Dataset contains student demographic and academic features
### GPA is used as a continuous target variable
### GradeClass represents categorized performance levels
### GPA distribution shows variation in student performance
### Dataset is suitable for both regression and classification tasks
