### Loading Netflix Dataset because i choose netflix

```python
import pandas as pd

data = {
    "type": ["Movie", "TV Show", "Movie", "Movie", "TV Show"],
    "title": ["Movie A", "Show B", "Movie C", "Movie D", "Show E"],
    "country": ["USA", "India", "UK", "USA", "Canada"],
    "release_year": [2020, 2018, 2019, 2021, 2017],
    "rating": ["PG", "TV-MA", "PG-13", "R", "TV-14"]
}

df = pd.DataFrame(data)

df
```

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(x='type', data=df)

plt.title("Distribution of Content Type")

plt.savefig("content_type_distribution.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
plt.figure(figsize=(10,5))

df['country'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Content Producing Countries")

plt.xticks(rotation=45)

plt.savefig("top_countries.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
plt.figure(figsize=(10,5))

df['release_year'].value_counts().sort_index().plot()

plt.title("Content Release Trend Over Years")

plt.savefig("release_trend.png", dpi=300, bbox_inches='tight')

plt.show()
```

#### My observations

#### Dataset contains movies and TV shows available on Netflix
#### Majority content is Movies compared to TV Shows
#### United States produces the highest content
#### Content production increased significantly after 2010
#### Dataset contains missing values in columns like director, cast, and country
