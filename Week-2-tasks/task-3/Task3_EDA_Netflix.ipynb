# Import & Load Cleaned Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

df = pd.read_csv("netflix_cleaned.csv")
```

### Movies vs TV Shows Distribution

```python
plt.figure(figsize=(6,4))

sns.countplot(x='type', data=df)

plt.title("Movies vs TV Shows Distribution")

plt.savefig("1_content_type.png", dpi=300, bbox_inches='tight')

plt.show()
```

### Top 10 Countries Producing Content

```python
plt.figure(figsize=(10,5))

df['country'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Content Producing Countries")

plt.xticks(rotation=45)

plt.savefig("2_top_countries.png", dpi=300, bbox_inches='tight')

plt.show()
```

### Content Added Over Time

```python
plt.figure(figsize=(10,5))

df['year_added'].value_counts().sort_index().plot()

plt.title("Content Added Over Years")

plt.savefig("3_year_added.png", dpi=300, bbox_inches='tight')

plt.show()
```

### Ratings Distribution

```python
plt.figure(figsize=(10,5))

sns.countplot(y='rating', data=df, order=df['rating'].value_counts().index)

plt.title("Content Ratings Distribution")

plt.savefig("4_ratings.png", dpi=300, bbox_inches='tight')

plt.show()
```

### Movie Duration Distribution

```python
# Extract only movies
movies = df[df['type'] == 'Movie']

# Extract numeric duration
movies['duration'] = movies['duration'].str.extract('(\d+)').astype(float)

plt.figure(figsize=(10,5))

sns.histplot(movies['duration'], bins=30)

plt.title("Movie Duration Distribution")

plt.savefig("5_duration.png", dpi=300, bbox_inches='tight')

plt.show()
```

### Correlation Heatmap

```python
plt.figure(figsize=(8,5))

numeric_df = df.select_dtypes(include=[np.number])

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')

plt.title("Correlation Heatmap")

plt.savefig("6_heatmap.png", dpi=300, bbox_inches='tight')

plt.show()
```

## Final Insights

### Netflix content is dominated by Movies over TV Shows
### United States contributes the highest amount of content
### Content addition increased significantly after 2015
### Most content falls under mature ratings like TV-MA
### Movie durations are mostly between 90–120 minutes
### Feature relationships were analyzed using correlation heatmap
