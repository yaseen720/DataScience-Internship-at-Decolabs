```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading Cleaned Dataset

```python
df = pd.read_csv("cleaned_dataset.csv")
df.head()
```

### Survival Count

```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

#### Survival by Gender

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.savefig("Survival by Gender.png", dpi=300)
plt.show()
```

#### Age Distribution

```python
sns.histplot(df['Age'], bins=30)
plt.title("Age Distribution")
plt.savefig("Age Distribution.png", dpi=300)
plt.show()
```

#### Passenger Class vs Survival

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Passenger Class")
plt.savefig("Survival by Passenger Class.png", dpi=300)
plt.show()
```

#### Fare Distribution

```python
sns.histplot(df['Fare'], bins=30)
plt.title("Fare Distribution")
plt.savefig("Fare Distribution.png", dpi=300)
plt.show()
```

#### Correlation Heatmap

```python
plt.figure(figsize=(10,6))
numeric_df = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.savefig("heatmap.png", dpi=300)  
plt.show()
```

## Insights

#### Majority passengers did not survive
#### Females had higher survival rate than males
#### Higher class passengers had better survival chances
#### Age distribution shows most passengers were between 20–40
