### Importing & Load Data

```python
import pandas as pd
import numpy as np

df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")   # same dataset
```

### checking missing values

```python
df.isnull().sum()
```

### now filling missing values

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

### fixing data format

```python
df['Age'] = df['Age'].astype(int)
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.rename(columns={'old_name': 'new_name'}, inplace=True)
```

```python
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

### checking outliers

```python
df.describe()
```

### saving cleaned dataset

```python
df.to_csv("cleaned_dataset.csv", index=False)
```
